```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("../fracture.csv")
df.sample(5)
```

```python
df = df.drop(columns=["id"])
df_cleanNaN = df.dropna()
dummi = pd.get_dummies(df_cleanNaN, drop_first=True)
dummi.sample(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dummi.drop(columns="fracture"), dummi["fracture"])
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
```

```python
pipeline = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=30)
)

pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.2%}")
```
